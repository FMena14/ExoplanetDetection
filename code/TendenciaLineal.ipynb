```python
import pandas as pd
import numpy as np

#df_l= pd.read_csv('Datos_split/koi_light_curves_train.csv', header=None)
df_train= pd.read_csv('Datos_split/koi_light_curves_train.csv', header=None)
df_val= pd.read_csv('Datos_split/koi_light_curves_val.csv', header=None)
df_test= pd.read_csv('Datos_split/koi_light_curves_test.csv', header=None)

df_train.shape
```

```python
dftrain_copia= df_train.copy()
dfval_copia= df_val.copy()
dftest_copia= df_test.copy()
```

```python
import time
start_time = time.time()
contador = 1
for line in df_train.values:
    if contador%250==0:
        print("Paso 250")
    puntos=0
    en_vacio=False
    line = np.nan_to_num(line)
    if np.max(line) == 0 and np.min(line)==0:
        #nada
        for i in range(0,len(line)):
            dftrain_copia.set_value(contador-1,i,0) 
        pass
    else:
        for indice in range(0, len(line)):
            if indice==0:
                if line[indice]==0:
                    a=0
                    ind_a=indice
                    en_vacio=True
                    puntos+=1
                else:
                    a=line[indice]
            elif indice==len(line)-1 and line[indice] == 0:
                if en_vacio:
                    b=0
                    den=indice-ind_a
                    m=(b-a)/den
                    for num in range(puntos+1):            
                        dftrain_copia.set_value(contador-1,indice-num,b-m*num)
                else:
                    dftrain_copia.set_value(contador-1,indice,0)
            else:
                if line[indice]==0 and en_vacio==False:
                    a=line[indice-1]
                    ind_a=indice-1
                    en_vacio=True
                    puntos+=1
                elif line[indice]==0 and en_vacio==True:
                    puntos+=1
                elif line[indice]!=0 and en_vacio==True:
                    en_vacio=False
                    b=line[indice]
                    den=indice-ind_a
                    m=(b-a)/den
                    for num in range(puntos+1):            
                        if num!=0:
                            dftrain_copia.set_value(contador-1,indice-num,b-m*num)
                    puntos=0
    contador+=1
print("Ha terminado")
```
