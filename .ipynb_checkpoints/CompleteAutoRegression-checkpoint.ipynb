```python
import matplotlib.pyplot as plt
import seaborn as sns
from astropy.table import Table
```

```python
print("Confirmed:")
mask = (df_label["NExScI Disposition"]=="CONFIRMED").values
conf_lc= lc[mask]
fig, ax = plt.subplots(10, 4, figsize = (12, 16))
for i in range(40):
    j = np.random.randint(0,conf_lc.shape[0])
    ax[i//4, i%4].plot(conf_lc[j])
    ax[i//4, i%4].get_xaxis().set_ticks([])
plt.show()
```

```python
print("False Positive:")
mask = (df_label["NExScI Disposition"]!="CONFIRMED").values
fp_lc= lc[mask]
fig, ax = plt.subplots(10, 4, figsize = (12, 16))
for i in range(40):
    j = np.random.randint(0,fp_lc.shape[0])
    ax[i//4, i%4].plot(fp_lc[j])
    ax[i//4, i%4].get_xaxis().set_ticks([])
plt.show()
```

```python
print("Estadisticos base ")
t = Table()
t["Type"] = ["Confirmed","False Positive"]
t["Minimum"] = [np.nanmin(conf_lc),np.nanmin(fp_lc)]
t["Maximum"] = [np.nanmax(conf_lc),np.nanmax(fp_lc)]
t["Mean"] = [np.nanmean(conf_lc),np.nanmean(fp_lc)]
t
```

```python
def autocorr(x):
    result = np.correlate(x, x, mode='full')
    return result[result.size/2:]

plt.plot(conf_lc[0])
plt.show()
plt.plot(autocorr(conf_lc[0]))
plt.show()
```

```python
#crearlas..
def create_dataset(secuencia,lag =1): 
    #preprocesamiento asumiendo dataset es una secuencia
    dataX = []
    dataY =[]
    for t_plus_1 in range(lag,len(secuencia)):
        if any(np.isnan( secuencia[t_plus_1-lag:t_plus_1+1])):
            continue
        #se crea el Y (target)
        predecir = secuencia[t_plus_1]
        dataY+= [predecir]
        
        #se crea el X (inputs) columnas para predecir Y
        columnasX = []
        for i in np.arange(lag,0,-1): #para los valores anteriores al t_plus_1 durante un lag
            columnasX += [secuencia[t_plus_1-i]]
        dataX.append(columnasX)
    return np.array(dataX),np.array(dataY)

def create_dataset_nan(secuencia,lag =1): 
    #preprocesamiento asumiendo dataset es una secuencia
    dataX = []
    dataY =[]
    for t_plus_1 in range(lag,len(secuencia)):
        if any(np.isnan( secuencia[t_plus_1-lag:t_plus_1+1])):
            continue
        #se crea el Y (target)
        predecir = secuencia[t_plus_1]
        dataY+= [predecir]
        
        #se crea el X (inputs) columnas para predecir Y
        columnasX = []
        for i in np.arange(lag,0,-1): #para los valores anteriores al t_plus_1 durante un lag
            columnasX += [secuencia[t_plus_1-i]]
        dataX.append(columnasX)
    return np.array(dataX),np.array(dataY)

A,B = create_dataset(conf_lc[0],3)
A
```
